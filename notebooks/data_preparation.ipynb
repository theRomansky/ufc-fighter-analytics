```python
import pandas as pd
import pickle
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../data/cleaned_data.csv')
```

```python
df.drop(['name'], axis=1, inplace=True)
```

```python
df = df.drop(['wins', 'losses', 'draws'], axis=1)
```

```python
df
```

```python
# Identify categorical and numerical features in the DataFrame
categorical_features = df.select_dtypes(include=['object']).columns.tolist()
numerical_features = df.select_dtypes(exclude=['object']).drop('win_percentage', axis=1).columns.tolist()
```

```python
# Standardize numerical features using StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[numerical_features])
df[numerical_features] = scaled_data
```

```python
# One-hot encode categorical features and replace them in the DataFrame
ohe = OneHotEncoder(sparse_output=False)
encoded_data = ohe.fit_transform(df[categorical_features])
df[ohe.get_feature_names_out()] = encoded_data
df.drop(categorical_features, axis=1, inplace=True)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df.drop(['win_percentage'], axis=1), df['win_percentage'], test_size=0.2, random_state=42)
```

```python
# Save the training and testing sets to a pickle file
with open('../data/datasets.pkl', 'wb') as f:
    pickle.dump((X_train, X_test, y_train, y_test), f)
```
